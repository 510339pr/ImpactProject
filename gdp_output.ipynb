```python
import pandas as pd
import numpy as np
from numpy import random 
import gdp_functions as gf
import itertools as it
```

```python
# Generate capabilities and alignment values 

# define start values and predefined knowledge 
stages = 10
scale = 10
# table amount of innovations, capabilities & alignment
inno = list(range(1,stages+1))
c = gf.gen_list_val(stages,scale)
a = gf.gen_list_val(stages,scale)

df = pd.DataFrame({'c':c, 'a':a}, index = inno)

# threshold values
C_agi = sum(c)*0.8 
A_aligned = sum(a)*0.8

# print the capabilities 
print('Capabilities vector:')
print(c)
print('Alignment vector')
print(a)
print('AGI threshold')
print(C_agi)
print('Alignment threshold')
print(A_aligned)
```

```python
# plot the innovations
gf.plot_innovations(inno, c, a, C_agi, A_aligned, False)
```

```python
# determine the optimal innovation path 
path = gf.optimal_path_rc(df, C_agi, A_aligned, stages)

# plot the optimal path
seq_c, seq_a, risk, aligned = gf.output_path(path, df, C_agi, A_aligned)
gf.plot_innovations(path, seq_c, seq_a, C_agi, A_aligned, True)
```
